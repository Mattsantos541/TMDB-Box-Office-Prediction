```python
#Beloings to collection new feature
#training / test split
X= df[['budget', 'popularity', 'runtime', 'belongs_to_collection' ]]
y= df['revenue']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .20, random_state = 123)

#X_train = X_train.reshape(-1, 1)
#X_test = X_test.reshape(-1, 1)

X_train.head()
```

```python
#model will be with linear regression target= revenue, input= budget, popularity, belongs_to Collection
lm1 = LinearRegression()

print(lm1)
```

```python
lm1.fit(X_train, y_train)
print("Linear Model:", lm1)

lm1_y_intercept = lm1.intercept_
print("intercept: ", lm1_y_intercept)

lm1_coefficients = lm1.coef_
print("coefficients: ", lm1_coefficients)
```

```python
#Insample predictions
y_pred_lm1 = lm1.predict(X_train)
```

```python
mse_lm1 = mean_squared_error(y_train, y_pred_lm1)
print("linear model\n  mean squared error: {:.3}".format(mse_lm1)) 

r2_lm1 = r2_score(y_train, y_pred_lm1)
print((r2_lm1))
```

```python
from math import sqrt
RMSE = sqrt(mse_lm1)

print("RMSE = ", RMSE)
```
